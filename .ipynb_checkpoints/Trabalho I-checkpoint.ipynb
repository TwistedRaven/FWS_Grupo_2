```python
import csv
import rdflib
from rdflib import Graph, Literal, Namespace, URIRef
#from rdflib.namespace import DCTERMS, RDF, RDFS, FOAF, XSD

NFL = Namespace('http://rdflib.net/test/')

dados_base = csv.DictReader(open("Basic_Stats_sample.csv"))
dados_passing = csv.DictReader(open("Career_Stats_Passing_sample.csv"))

```

```python
dados = Graph()
```

```python
for row in dados_base:
    id =row['Player Id'] 
    jogador = URIRef("http://rdflib.net/test/id/"+id)

    dados1.add((jogador, NFL['id'], jogador))
    dados1.add((jogador, NFL['name'], Literal(row['Name'] )))
    dados1.add((jogador, NFL['Team'], Literal(row['Current Team'] )))
    dados1.add((jogador, NFL['position'], Literal(row['Position'] )))
    dados1.add((jogador, NFL['player_number'], Literal(row['Number'] )))
    dados1.add((jogador, NFL['Current_Status'], Literal(row['Current Status'] )))
    dados1.add((jogador, NFL['Experience'], Literal(row['Experience'] )))
    dados1.add((jogador, NFL['Birth_Place'], Literal(row['Birth Place'] )))
    dados1.add((jogador, NFL['College'], Literal(row['College'] )))
    dados1.add((jogador, NFL['High_School'], Literal(row['High School'] )))
    dados1.add((jogador, NFL['High_School_Location'], Literal(row['High School Location'] ))) 
    dados1.add((jogador, NFL['Height'], Literal(row['Height (inches)'] )))
    dados1.add((jogador, NFL['Weight'], Literal(row['Weight (lbs)'] )))
    dados1.add((jogador, NFL['Years_Played'], Literal(row['Years Played'] )))
    
    
   
```
